```python
from sqlalchemy import create_engine, text
import pandas as pd
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')

engine = create_engine('postgresql://weather_user:weather_pass@localhost:5432/weather')
print('Engine créé ✅')
```

```python
# Test connexion
with engine.connect() as conn:
    version = conn.execute(text('SELECT version()')).fetchone()[0]
print('✅ Connecté à PostgreSQL')
print(version)
```

```python
# Compter les lignes
with engine.connect() as conn:
    count = conn.execute(text('SELECT COUNT(*) FROM current_weather')).fetchone()[0]
print(f'📊 Villes en base : {count}')
```

```python
# Météo actuelle
with engine.connect() as conn:
    current = pd.read_sql(text("""
        SELECT city, country, temperature_c, apparent_temperature_c,
               humidity_pct, wind_speed_kmh, wind_gusts_kmh,
               precipitation_mm, pressure_hpa,
               weather_description, alert_level, timestamp
        FROM current_weather
        ORDER BY city
    """), conn)

print(f'{len(current)} villes')
display(current)
```

```python
# Stats globales
with engine.connect() as conn:
    stats = pd.read_sql(text("""
        SELECT
            COUNT(*)                                  AS nb_villes,
            ROUND(AVG(temperature_c)::numeric, 1)     AS temp_moy_c,
            ROUND(MIN(temperature_c)::numeric, 1)     AS temp_min_c,
            ROUND(MAX(temperature_c)::numeric, 1)     AS temp_max_c,
            ROUND(AVG(humidity_pct)::numeric, 1)      AS humidite_moy,
            ROUND(MAX(wind_speed_kmh)::numeric, 1)    AS vent_max_kmh,
            SUM(CASE WHEN alert_level != 'normal' THEN 1 ELSE 0 END) AS alertes
        FROM current_weather
    """), conn)
display(stats)
```

```python
# Historique 6h
with engine.connect() as conn:
    history = pd.read_sql(text("""
        SELECT city, temperature_c, humidity_pct,
               wind_speed_kmh, alert_level, timestamp
        FROM weather_history
        WHERE timestamp >= NOW() - INTERVAL '6 hours'
        ORDER BY timestamp ASC
    """), conn)

print(f'Historique : {len(history)} points')
display(history.head(10))
```

```python
# Graphique température par ville
fig = px.bar(
    current.sort_values('temperature_c'),
    x='temperature_c', y='city', orientation='h',
    color='temperature_c', color_continuous_scale='RdBu_r',
    title='Température actuelle par ville (°C)', height=600
)
fig.show()
```

```python
# Évolution température dans le temps
if not history.empty:
    fig = px.line(
        history, x='timestamp', y='temperature_c', color='city',
        title='Évolution des températures (6h)', height=500
    )
    fig.update_layout(hovermode='x unified')
    fig.show()
else:
    print('Pas encore assez de données historiques.')
```

```python
# Alertes
with engine.connect() as conn:
    alerts = pd.read_sql(text("""
        SELECT city, alert_level, alert_message, timestamp
        FROM weather_alerts
        ORDER BY timestamp DESC LIMIT 50
    """), conn)

if alerts.empty:
    print('✅ Aucune alerte active.')
else:
    print(f'⚠️ {len(alerts)} alertes :')
    display(alerts)
```
